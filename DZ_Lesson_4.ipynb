```python
import pandas as pd
import matplotlib.pyplot as pit
import seaborn as sns
```

```python
data = pd.read_csv('kc_house_data.csv', encoding='Windows-1251')
data.head()
```

```python
pit.figure(figsize= (6,4))
pit.hist(data['price'], bins=15)
pit.title('Стоимость домов')
pit.xlabel('USD')
pit.ylabel('Количество домов')
```

* Изучите стоимости недвижимости

```python
pit.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True)
pit.title('Гистограмма цен на недвижимость')
pit.xlabel('Цена')
pit.ylabel('Частота')
pit.show()
```

*  Изучите распределение квадратуры жилой

```python
pit.figure(figsize=(10, 6))
sns.histplot(data['sqft_living'], kde=True)
pit.title('Распределение квадратуры жилой площади')
pit.xlabel('Квадратура жилой площади')
pit.ylabel('Частота')
pit.show()
```

```python
pit.figure(figsize=(10, 6))
pit.hist(data['yr_built'], bins=20, edgecolor='k')
pit.title('Распределение года постройки')
pit.xlabel('Год постройки')
pit.ylabel('Частота')
pit.show()
```

# Условие 2: 2 задача

* Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
waterfront_counts = data['waterfront'].value_counts()

pit.figure(figsize=(8, 6))
sns.barplot(x=waterfront_counts.index, y=waterfront_counts.values)
pit.title('Распределение домов относительно наличия вида на набережную')
pit.xlabel('Вид на набережную')
pit.ylabel('Количество домов')
pit.xticks([0, 1], ['Нет вида', 'Есть вид'])
pit.show()
```

* Изучите распределение этажей домов

```python
pit.figure(figsize=(10, 6))
pit.hist(data['floors'], bins=20, edgecolor='k')
pit.title('Распределение этажей в домах')
pit.xlabel('Количество этажей')
pit.ylabel('Частота')
pit.show()
```

* Изучите распределение состояния домов

```python
condition_counts = data['view'].value_counts()

pit.figure(figsize=(8, 6))
sns.barplot(x=condition_counts.index, y=condition_counts.values)
pit.title('Распределение состояния домов')
pit.xlabel('Состояние')
pit.ylabel('Количество домов')
pit.show()
```
